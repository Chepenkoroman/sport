```python
import pandas as pd
```

```python
data = pd.read_csv('spi_matches.csv')
data.head()
```

```python
data['league'].value_counts()
```

```python
dataUCL = data.loc[data['league'] == 'UEFA Champions League']
dataUCL.head()
```

```python
dataUCL['season'].value_counts()
```

```python
dataUCL22 = dataUCL.loc[dataUCL['season'] == 2022]
dataUCL22.head()
```

```python
dataUCL22['date'] = pd.to_datetime(data['date'])
dataUCL22['Year'] = pd.to_datetime(data['date']).dt.year
dataUCL22['Month'] = pd.to_datetime(data['date']).dt.month
dataUCL22.head()
```

```python
dataUCL['date'] = pd.to_datetime(dataUCL['date'])
dataUCL['Year'] = pd.to_datetime(dataUCL['date']).dt.year
dataUCL['Month'] = pd.to_datetime(dataUCL['date']).dt.month
dataUCL['Day'] = pd.to_datetime(dataUCL['date']).dt.day
dataUCL.head()
```

```python
dataUCL.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
# Load the example flights dataset and convert to long-form

ucl = dataUCL.pivot_table(index="Year", columns="Month", values="xg1", aggfunc='sum')
```

```python
# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(ucl, annot=True, fmt="d", linewidths=.5, ax=ax)
```
